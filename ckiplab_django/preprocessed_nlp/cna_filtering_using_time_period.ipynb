## Find news with a specified date

### Using query command

```python
df.query("date == '2022-09-29'")
```

### Using column value

```python
df[df['date'] == '2022-09-29']
```

```python
# number of pieces
len(df[df['date'] == '2022-09-29'])
```

### Find news with date greater than the start date and less than the end date

```python
df_interval = df[(df['date'] >= '2022-09-23') & (df['date'] <= '2022-09-29')]
len(df_interval )
```

```python
# same result using query
df_interval = df.query("date >= '2022-09-23' & date <= '2022-09-29'")
len(df_interval )
```

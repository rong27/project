# ckiplab word segment (中研院斷詞)

```python
from ckiptagger import WS, POS, NER
```

```python
# location of models
model_path = 'ckiplab-model-data/data/'

ws = WS(model_path)
pos = POS(model_path)
ner = NER(model_path)
```

## Word Segmentation

```python
%%time

tokens = ws(df.content)
```

```python
len(tokens)
```

```python
tokens[0]
```

## POS 詞性標註

```python
%%time
tokens_pos = pos(tokens)
```

```python
len(tokens_pos)
```

```python
tokens_pos[0]
```

## word pos pair 斷詞語詞性配對

```python
word_pos_pair = [list(zip(w, p)) for w, p in zip(tokens, tokens_pos)]
```

```python
len(word_pos_pair)
```

```python
word_pos_pair[0][0]
```

## NER命名實體辨識，如數字、日詞、事件、行政區...,etc

```python
%%time
entity_list = ner(tokens, tokens_pos)
```

```python
entity_list[0]  

## (0, 3, 'ORG', '外交部') 從0的位置開始到3 = for i in range(0, 3)
```

# Remove stop words and filter using POS (tokens_rm_stopwords)

去除停用詞並依據詞性過濾

```python
import os
os.getcwd()
```

```python
with open('stops_chinese_traditional.txt', 'r', encoding='utf8') as f:
    stops = f.read().split('\n') 
```

```python
# 過濾條件:兩個字以上與特定的詞性
allowPOS=['Na','Nb','Nc','VA','VAC','VB','VC', 'A']

tokens_rm_stopwords =[]
for seg_pos in word_pos_pair:
    tokens_rm_stopwords.append([w for w, p in seg_pos if w not in stops and (len(w) >= 2) and p in allowPOS])
```

```python
tokens_rm_stopwords[0]
```

# Calculate word count (frequency) 計算字頻(次數)

```python
# allowPOS過濾條件:兩個字以上與特定的詞性
allowPOS = ['Na','Nb','Nc','VA','VAC','VB','VC']

from collections import Counter
def word_frequency(wp_pair):
    filtered_words = []
    for word, pos in wp_pair:
        if (pos in allowPOS) and (len(word) >= 2):
            filtered_words.append(word)
#         print('%s %s' % (word, pos))
    counter = Counter(filtered_words)
    return counter.most_common(200)
```

```python
word_frequency(word_pos_pair[0])
```

```python
%%time
keyfreqs = []
for wp in word_pos_pair:
    topwords = word_frequency(wp)
    keyfreqs.append(topwords)  
```

```python
keyfreqs[0:1]
```

```python
df['top_key_freq'] = keyfreqs
```

### Sentimental analysis - step1. 取出重要關鍵詞

### Sentimental analysis - step3. 計算情緒分數

```python
sn.sentiments
```

```python
%%time
summary = []
sentiment = []
for text in df.content: # process piece by piece
    sn = SnowNLP(text)
    summary.append(sn.summary())  
    sentiment.append(round(sn.sentiments, 2))
```

```python
df['summary'] = summary
```

```python
df['sentiment'] = sentiment
```
